```python
import pandas as pd
import numpy as np
```

```python
sessions_df = pd.read_csv("cinema_sessions.csv", sep=' ')
titanic_df = pd.read_csv("titanic_with_labels.csv", sep=' ')
display(sessions_df.head())
display(titanic_df.head())
```

```python
titanic_df.dtypes
```

```python
titanic_df.dropna(subset=["sex"])
titanic_df['sex'] = titanic_df['sex'].str.lower()
titanic_df['sex'] = titanic_df['sex'].map({'м' : 0, 'ж' : 1, 'm' : 0})
titanic_df['row_number'] = titanic_df['row_number'].fillna(titanic_df['row_number'].max())

display(titanic_df.head())
```

```python
Q1 = titanic_df['liters_drunk'].quantile(0.25)
Q3 = titanic_df['liters_drunk'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
mean_liters = titanic_df[(titanic_df['liters_drunk'] >= 0) & 
                                      (titanic_df['liters_drunk'] <= upper_bound)]['liters_drunk'].mean()
titanic_df['liters_drunk'] = titanic_df['liters_drunk'].apply(
    lambda x: mean_liters if x < 0 or x > upper_bound else x)

print(f"Mean liters: {mean_liters}")
display(titanic_df.head())
```
